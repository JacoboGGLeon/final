```python
file = "../../../../../../dataframes/enoe/SDEMT/df_09_impute.csv"
```

```python
df = pd.read_csv(file)
```

```python
df.head(1)
```

```python
df.shape
```

```python
df_impute = df.copy()
```

```python
df = df_impute
```

```python
df[["CVE_LOC", "CVE_MUN"]] = df_clean
```

```python
df.head(1)
```

```python
df_copy = df.copy()
```

```python
df.shape
```

```python
list_num = ["eda", "anios_esc", "hrsocup", "ingocup", "ing_x_hrs"]
```

```python
list_cat = []

for row in df_copy.columns:
    if row not in list_num:
#            print(row)
        list_cat.append(row)
            
list_cat
```

```python
list_cat.remove('CVE_LOC')
```

```python
list_cat.remove('CVE_MUN')
```

```python
f = {}

for column in df_copy.columns:
#    print(column)
    if(column in df_copy[list_cat].columns):
#        print(column, 'categoricos')
        f[column] = lambda x: x.value_counts().index[0]
    elif(column in df_copy[list_num].columns):
#        print(column, 'numericos')
        f[column] = 'mean'
    
f
```

```python
grouped = df_copy.groupby(by=['CVE_MUN', 'CVE_LOC']).agg(f)
grouped
```

```python
grouped.shape
```

```python
file = "../../../../../../dataframes/enoe/SDEMT/df_grouped.csv"
```

```python
grouped.to_csv(file,index = False, header = True, sep = ',', encoding = 'utf-8')
```
