```python
import pandas as pd
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = "../../../../../../dataframes/enh/df_09.csv"
```

```python
df = pd.read_csv(file)
```

```python
df.shape
```

```python
df.head()
```

```python
df_copy = df.copy()
```

```python
df_copy.replace(to_replace=' ', value=np.NaN, inplace=True)
```

```python
df_copy.replace(to_replace='&', value=np.NaN, inplace=True)
```

```python
df_copy.replace(to_replace=np.NaN, value=-1, inplace=True)
```

```python
df_copy = df_copy.astype('int64')
```

```python
l = []
bandera = np.NAN
for column in df_copy.columns:
    if df[column].dtype == np.int64:
        bandera = 'entero'
    elif df[column].dtype == np.float64:
        bandera = 'flotante'
    else:
        bandera = 'otro'
        
    print('COLUMNA: ', column)
    print('TIPO: ', bandera)
    print('NUNIQUE: ', df_copy[column].nunique())
    print('UNIQUE: ', df_copy[column].unique())
    print('VALUE COUNTS ', df_copy[column].value_counts())
    print('\n')
```

```python
df_copy.replace(to_replace=-1, value=np.NaN, inplace=True)
```

```python
df_copy.head()
```

```python
msno.matrix(df_copy, fontsize=50, sparkline=False, labels=False)
plt.xlabel('Columnas', fontsize=20)
plt.ylabel('Observaciones', fontsize=20)
#plt.savefig('../../../images/envipe/df_tmod_vic_envipe2016_09.png', dpi=300)
```

```python
msno.bar(df_copy, fontsize=10, labels=False)
plt.xlabel('Columnas', fontsize=20)
plt.ylabel('Observaciones', fontsize=20)
#plt.savefig('../../../images/envipe/df_tmod_vic_envipe2016_09.png', dpi=300)
```

```python
df_copy.replace(to_replace=np.NaN, value=-1, inplace=True)
```

```python
df_copy.columns[df_copy.isna().any()].tolist()
```

```python
threshold = round(df_copy.shape[0]*.8)
threshold
```

```python
l = []
for column in df_copy.columns:
    na_column =  df_copy[column].value_counts().get(-1)
    if(na_column != None):
        if (df_copy.shape[0] - na_column < threshold):
            print(column, ': ', df_copy.shape[0] - na_column, ' - ', threshold, 'd e l e t e d')
            l.append(column)
```

```python
l.remove('antiguedad')
```

```python
df_copy.drop(labels=l, axis=1, inplace=True)
```

```python
df_copy.fillna(value=-1, inplace=True)
```

```python
df_copy.replace(to_replace=-1, value=np.NaN, inplace=True)
```

```python
msno.bar(df_copy, fontsize=10, labels=False)
plt.xlabel('Columnas', fontsize=20)
plt.ylabel('Observaciones', fontsize=20)
#plt.savefig('../../../images/envipe/df_tmod_vic_envipe2016_09.png', dpi=300)
```

```python
file = "../../../../../../dataframes/enh/df_09_clean.csv"
```

```python
df_copy.to_csv(file,index = False, header = True, sep = ',', encoding = 'utf-8')
```
