```python
list_numericos = []

for row in df_copy.columns:
    if row not in list_cat:
#        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
        if (row != 'CVE_MUN') & (row != 'cluster'):
#            print(row)
            list_numericos.append(row)
            
list_numericos
```

```python
f = {}

for column in df_copy.columns:
#    print(column)
    if(column in df_copy[list_cat].columns):
#        print(column, 'categoricos')
        f[column] = lambda x: x.value_counts().index[0]
    elif(column in df_copy[list_numericos].columns):
#        print(column, 'numericos')
        f[column] = 'mean'
    
f
```

```python
grouped = pd.DataFrame(df_copy.groupby(by=['cluster']).agg(f))
grouped.reset_index(inplace=True)
grouped
```
