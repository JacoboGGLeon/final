```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
file = "../../../../../dataframes/ensu/ENSU_CB/df_09_impute.csv"
```

```python
df = pd.read_csv(file)
```

```python
df.shape
```

```python
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
dataframe_to_transform = df_copy.copy()

for column in df_copy.columns:
    df = pd.DataFrame(dataframe_to_transform[column].value_counts(normalize=True))
    #df['cumsum']=(np.cumsum(df[column]))/2
    df['cumsum']=np.cumsum(df[column])
#    print(df)
    
    if df[df.duplicated(['cumsum'], keep=False)].any() is True:
        print(df[df.duplicated(['cumsum'], keep=False)])
    else:
        dict_to_transform = df['cumsum'].to_dict()
#        print(dict_to_transform)
        dataframe_to_transform[column] = dataframe_to_transform[column].replace(dict_to_transform)
#        print(dataframe_to_transform)
```

```python
dataframe_to_transform.head()
```

```python
df_copy = dataframe_to_transform
df_copy.head()
```

```python
list_regejos = df_copy.columns[df_copy.isnull().any()].tolist()
list_regejos
```

```python
df_copy.drop(labels=list_regejos, axis=1, inplace=True)
```

```python
file = "../../../../../dataframes/ensu/ENSU_CB/df_09_transform.csv"
```

```python
df_copy.to_csv(file,index = False, header = True, sep = ',', encoding = 'utf-8')
```

## ONE HOT ENCODING

```python
file = "../../../../../dataframes/ensu/ENSU_CB/df_09_impute.csv"
```

```python
df = pd.read_csv(file)
```

```python
df.head(2)
```

```python
df_copy = df.copy()
```

```python
df_copy.head(2)
```

```python
list_cat = df_copy.columns
```

```python
from numpy import array
from numpy import argmax
from keras.utils import to_categorical
```

```python
for column in df_copy[list_cat].columns:
    print(column, df_copy[column].unique())
    encoded = to_categorical(y = df_copy[column], dtype='int64')
#    print(encoded.shape)
    df_encoded = pd.DataFrame(encoded)
#    print(df_encoded.shape)

    df_copy.drop(labels=[column], axis=1, inplace=True)

    for element in list(df_encoded.columns):
        var = column + str("_") + str(element)
#        print(var)
        df_copy[var] = df_encoded[element]
#        print(df_copy[var].head(2))
```

```python
df_copy.shape
```

```python
list_zeros = []

for column in df_copy.columns:
    if df_copy[column].nunique() == 1 :
        list_zeros.append(column)
```

```python
df_copy.drop(labels=list_zeros, axis=1, inplace=True)
```

```python
df_copy.shape
```

```python
file = "../../../../../dataframes/ensu/ENSU_CB/df_encoding.csv"
```

```python
df_copy.shape
```

```python
df_copy.to_csv(file,index = False, header = True, sep = ',', encoding = 'utf-8')
```
