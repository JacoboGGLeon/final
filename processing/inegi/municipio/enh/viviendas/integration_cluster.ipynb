## SUMMARY

```python
df_copy = df_vect_best_entropy_copy.copy()
df_copy.head(2)
```

```python
list_numericos = ['antiguedad','cuart_dorm','bano_comp','bano_excus', 'bano_regad','focos_ahor', 'focos_inca','num_cuarto']
```

```python
list_cat = []

for row in df_copy.columns:
    if row not in list_numericos:
#        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
        if (row != 'CVE_MUN'):
#            print(row)
            list_cat.append(row)
            
list_cat
```

```python
f = {}

for column in df_copy.columns:
#    print(column)
    if(column in df_copy[list_cat].columns):
#        print(column, 'categoricos')
        f[column] = lambda x: x.value_counts().index[0]
    elif(column in df_copy[list_numericos].columns):
#        print(column, 'numericos')
        f[column] = 'mean'
    
f
```

```python
grouped = pd.DataFrame(df_copy.groupby(by=['CVE_MUN']).agg(f))
grouped.reset_index(inplace=True)
grouped
```

```python
grouped[["CVE_MUN", "cluster"]]
```

# POR CLUSTER

```python
list_numericos = ['antiguedad','cuart_dorm','bano_comp','bano_excus', 'bano_regad','focos_ahor', 'focos_inca','num_cuarto']
```

```python
list_cat = []

for row in df_copy.columns:
    if row not in list_numericos:
#        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
        if (row != 'CVE_MUN') & (row != 'cluster'):
#            print(row)
            list_cat.append(row)
            
list_cat
```

```python
f = {}

for column in df_copy.columns:
#    print(column)
    if(column in df_copy[list_cat].columns):
#        print(column, 'categoricos')
        f[column] = lambda x: x.value_counts().index[0]
    elif(column in df_copy[list_numericos].columns):
#        print(column, 'numericos')
        f[column] = 'median'
    
f
```

```python
pd.set_option('display.max_columns', 500)
```

```python
grouped = pd.DataFrame(df_copy.groupby(by=['cluster']).agg(f))
grouped.reset_index(inplace=True)
grouped
```
