```python
import pandas as pd
import numpy as np
```

```python
file = '../../../../../../dataframes/enigh/HOGARES/df_09_impute.csv'
```

```python
df = pd.read_csv(file)
```

```python
df.shape
```

```python
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
list_cat = ["telefono", "celular", "tv_paga", "conex_inte", "esc_radio", "tarjeta", "pagotarjet", "regalotar", "regalodado", "autocons", "regalos", "remunera", "transferen", "parto_g", "embarazo_g", "negcua", "bene_licon", "diconsa", "acc_alim1", "acc_alim2", "acc_alim3", "acc_alim4", "acc_alim5", "acc_alim6", 'num_trab_d', 'num_auto', 'num_van', 'num_pickup', 'num_moto', 'num_bici', 'num_trici','num_otro', 'num_ester', 'num_grab', 'num_radio', 'num_tva', 'num_tvd', 'num_dvd', 'num_video', 'num_licua', 'num_tosta', 'num_micro', 'num_refri', 'num_estuf', 'num_lavad', 'num_planc', 'num_maqui', 'num_venti', 'num_aspir', 'num_compu', 'num_impre', 'num_juego', 'alim17_1', 'alim17_2', 'alim17_3', 'alim17_4', 'alim17_5', 'alim17_6', 'alim17_7', 'alim17_8', 'alim17_9', 'alim17_10', 'alim17_11', 'alim17_12', 'acc_alim18']
```

```python
list_numericos = []

for row in df_copy.columns:
    if row not in list_cat:
#        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
        if (row != 'mun'):
#            print(row)
            list_numericos.append(row)
            
list_numericos
```

```python
df_copy[list_cat].head()
```

```python
dataframe_to_transform = df_copy[list_cat].copy()

for column in df_copy[list_cat].columns:
#    df = pd.DataFrame(dataframe_to_transform[column].value_counts(normalize=True, sort=False, ascending=True))
    df = pd.DataFrame(dataframe_to_transform[column].value_counts(normalize=True))
#    print(df)
#    df['cumsum']=(np.cumsum(df[column]))/10
    df['cumsum']=np.cumsum(df[column])
#    print(df)
    
    if df[df.duplicated(['cumsum'], keep=False)].any() is True:
        print(df[df.duplicated(['cumsum'], keep=False)])
    else:
        dict_to_transform = df['cumsum'].to_dict()
#        print(dict_to_transform)
        dataframe_to_transform[column] = dataframe_to_transform[column].replace(dict_to_transform)
#        print(dataframe_to_transform)
```

```python
dataframe_to_transform.head()
```

```python
df_num = df_copy[list_numericos]
dataframe_to_normalize=((df_num-df_num.min())/(df_num.max()-df_num.min()))*1
```

```python
dataframe_to_normalize.head()
```

```python
df_copy[list_cat] = dataframe_to_transform
df_copy[list_numericos] = dataframe_to_normalize
df_copy.head()
```

## ONE HOT ENCODING

```python
file = '../../../../../../dataframes/enigh/HOGARES/df_09_impute.csv'
```

```python
df = pd.read_csv(file)
```

```python
from numpy import array
from numpy import argmax
from keras.utils import to_categorical
```

```python
for column in df_copy[list_cat].columns:
    print(column, df_copy[column].unique())
    encoded = to_categorical(y = df_copy[column], dtype='int64')
#    print(encoded.shape)
    df_encoded = pd.DataFrame(encoded)
#    print(df_encoded.shape)

    df_copy.drop(labels=[column], axis=1, inplace=True)

    for element in list(df_encoded.columns):
        var = column + str("_") + str(element)
#        print(var)
        df_copy[var] = df_encoded[element]
#        print(df_copy[var].head(2))
```

```python
df_copy.shape
```

```python
list_zeros = []

for column in df_copy.columns:
    if df_copy[column].nunique() == 1 :
        list_zeros.append(column)
```

```python
df_copy.drop(labels=list_zeros, axis=1, inplace=True)
```
