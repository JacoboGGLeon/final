```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
file = '../../../../../../dataframes/enigh/HOGARES/df_09_impute.csv'
```

```python
df = pd.read_csv(file)
```

```python
df.shape
```

```python
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
df_copy.drop(labels='huespedes', axis=1, inplace=True)
```

```python
list_cat = ["telefono", "celular", "tv_paga", "conex_inte", "esc_radio", "tarjeta", "pagotarjet", "regalotar", "regalodado", "autocons", "regalos", "remunera", "transferen", "parto_g", "embarazo_g", "negcua", "bene_licon", "diconsa", "acc_alim1", "acc_alim2", "acc_alim3", "acc_alim4", "acc_alim5", "acc_alim6"]
```

```python
list_numericos = []

for row in df_copy.columns:
    if row not in list_cat:
#        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
        if (row != 'mun'):
#            print(row)
            list_numericos.append(row)
            
#list_numericos
```

```python
df_copy[list_cat].head()
```

```python
dataframe_to_transform = df_copy[list_cat].copy()

for column in df_copy[list_cat].columns:
    df = pd.DataFrame(dataframe_to_transform[column].value_counts(normalize=True, sort=False, ascending=True))
#    df['cumsum']=(np.cumsum(df[column]))/10
    df['cumsum']=np.cumsum(df[column])
#    print(df)
    
    if df[df.duplicated(['cumsum'], keep=False)].any() is True:
        print(df[df.duplicated(['cumsum'], keep=False)])
    else:
        dict_to_transform = df['cumsum'].to_dict()
#        print(dict_to_transform)
        dataframe_to_transform[column] = dataframe_to_transform[column].replace(dict_to_transform)
#        print(dataframe_to_transform)
```

```python
dataframe_to_transform.head()
```

```python
df_num = df_copy[list_numericos]
dataframe_to_normalize=((df_num-df_num.min())/(df_num.max()-df_num.min()))*1
```

```python
dataframe_to_normalize.head()
```

```python
df_copy[list_cat] = dataframe_to_transform
df_copy[list_numericos] = dataframe_to_normalize
df_copy.head()
```

```python
list_regejos = df_copy.columns[df_copy.isnull().any()].tolist()
list_regejos
```

```python
df_copy.drop(labels=list_regejos, axis=1, inplace=True)
```

```python
file = '../../../../../../dataframes/enigh/HOGARES/df_transform_cat_to_num.csv'
```

```python
df_copy.to_csv(file,index = False, header = True, sep = ',', encoding = 'utf-8')
```
