```python
list_text = ['BP1_5_5E','BP2_2_17E', 'BP2_3_6E', 'BP2_3_7E', 'BP2_4_11E', 'BP2_4_9E', 'BP3_1_14E', 'BP3_1_16E',
 'DOMINIO','FAC_SEL', 'NOM_ENT', 'NOM_MUN', 'SEL_SUST', 'TIPO_CUEST', 'UPM', 'UPM_DIS']
df.drop(labels=list_text, axis=1, inplace=True)
```

```python
df.replace(to_replace='\r',value=np.NaN,inplace=True)
```

```python
df.replace(to_replace='-\r',value=np.NaN,inplace=True)
```

```python
df.replace({'1\r':1,  
           '2\r':2,
           '3\r':3,
           '4\r':4,
           '5\r':5,
           '6\r':6,
           '7\r':7,
           '8\r':8,
           '9\r':9,
           '0\r':0},
           inplace=True)
```

```python
l = []
bandera = np.NAN
for column in df.columns:
    try:
        df[column].astype('int64')
    except:
        l.append(column)
    
    if df[column].dtype == np.int64:
        bandera = 'entero'
    elif df[column].dtype == np.float64:
        bandera = 'flotante'
    else:
        bandera = 'otro'
        
    print('COLUMNA: ', column)
    print('TIPO: ', bandera)
    print('NUNIQUE: ', df[column].nunique())
    print('UNIQUE: ', df[column].unique())
    print('VALUE COUNTS ', df[column].value_counts())
    print('\n')
```
