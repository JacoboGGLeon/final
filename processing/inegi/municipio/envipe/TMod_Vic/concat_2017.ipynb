### Vivienda

```python
file_tvivienda_envipe2016 = '../../../../../../data/inegi-encuestas/municipio/envipe/envipe_2017_csv/tvivienda_envipe2017/conjunto_de_datos/tvivienda.csv'
```

```python
df_tvivienda_envipe2016 = pd.read_csv(file_tvivienda_envipe2016, encoding='latin')
```

```python
df_tvivienda_envipe2016_copy = df_tvivienda_envipe2016.copy()
```

```python
df_tvivienda_envipe2016_copy.rename(columns={"ENT":"CVE_ENT", "MUN":"CVE_MUN"}, inplace=True)
```

```python
df_tvivienda_envipe2016_copy.head()
```

```python
df_merge_tmod_vic_envipe2016_copy = pd.merge(left=df_01, 
                                             right=df_tvivienda_envipe2016_copy[['ID_VIV', 'CVE_ENT', 'CVE_MUN']],
                                             how='inner', 
                                             on=['ID_VIV'])
```

```python
df_merge_tmod_vic_envipe2016_copy.shape
```

```python
df_merge_tmod_vic_envipe2016_copy.columns
```

### Subset Ciudad de México

```python
df_tmod_vic_envipe2016_09 = df_merge_tmod_vic_envipe2016_copy[df_merge_tmod_vic_envipe2016_copy.CVE_ENT == 9]
```

```python
df_tmod_vic_envipe2016_09.head()
```

```python
df_tmod_vic_envipe2016_09.shape
```

```python
df_copy=df_tmod_vic_envipe2016_09.copy()
```

```python
#list(df_copy.columns)
```

```python
list_identificadores = ['ID_VIV', 'ID_HOG', 'ID_PER', 'ID_DEL', 'UPM', 'VIV_SEL', 'HOGAR', 'ND_TIPO', 'TD_TIPO','R_SEL', 'AREAM_OCU', 'RESUL_H','FAC_DEL', 'FAC_DEL_AM', 'DOMINIO','EST_DIS', 'UPM_DIS']
```

```python
df_copy.drop(labels=list_identificadores, axis=1, inplace=True)
```

```python
df_copy.replace(to_replace='\r',value=np.NaN,inplace=True)
```

```python
df_copy.replace({'0\r':0,
                 '1\r':1,
                 '2\r':2,
                 '3\r':3,
                 '4\r':4,
                 '5\r':5,
                 '6\r':6,
                 '7\r':7,
                 '8\r':8,
                 '9\r':9,
                 '99\r':99}, inplace=True)
```

```python
df_copy.replace({'00\r':0,
                 '01\r':1,
                 '02\r':2,
                 '03\r':3,
                 '04\r':4,
                 '05\r':5,
                 '06\r':6,
                 '07\r':7,
                 '08\r':8,
                 '09\r':9,
                 '10\r':10,
                 '11\r':11,
                 '12\r':12,
                 '13\r':13,
                 '14\r':14,
                 '15\r':15,
                 '99\r':99,}, inplace=True)
```

```python
df_copy.replace(to_replace=np.NaN,value=-1,inplace=True)
```

```python
l = []
bandera = np.NAN
for column in df_copy.columns:
    if df_copy[column].dtype == np.int64:
        bandera = 'entero'
    elif df_copy[column].dtype == np.float64:
        bandera = 'flotante'
    else:
        bandera = 'otro'
    
    try:
        df_copy[column] = df_copy[column].astype('int64')
    except:
        #pass#
        l.append(column)
    
    print('COLUMNA: ', column)
    print('TIPO: ', bandera)
    print('NUNIQUE: ', df_copy[column].nunique())
    print('UNIQUE: ', df_copy[column].unique())
    print('VALUE COUNTS ', df_copy[column].value_counts())
    print('\n')
```
