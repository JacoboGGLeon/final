```python
import pandas as pd
import numpy as np
import os
```

```python
file_tmod_vic_envipe2016_09_impute = '../../../../dataframes/envipe/2016/tmod_vic_envipe2016_09_impute.csv'
```

```python
df_tmod_vic_envipe2016_09_impute = pd.read_csv(file_tmod_vic_envipe2016_09_impute)
df_tmod_vic_envipe2016_09_impute.head()
```

```python
df_tmod_vic_envipe2016_09_impute_copy = df_tmod_vic_envipe2016_09_impute.copy()
```

```python
df_tmod_vic_envipe2016_09_impute_copy.BP1_2C.value_counts(sort=False, ascending=True)
```

```python
df = pd.DataFrame(df_tmod_vic_envipe2016_09_impute_copy.BP1_2C.value_counts(normalize=True, sort=False, ascending=True))
df.reset_index(inplace=True)
df
```

```python
(np.cumsum(df.BP1_2C))/df.BP1_2C.nunique()
```

```python
dataframe_to_transform = df_tmod_vic_envipe2016_09_impute_copy.copy()

for column in df_tmod_vic_envipe2016_09_impute_copy.columns:
    df = pd.DataFrame(dataframe_to_transform[column].value_counts(normalize=True, sort=False, ascending=True))
    df['cumsum']=(np.cumsum(df[column]))/10
#    print(df)
    
    if df[df.duplicated(['cumsum'], keep=False)].any() is True:
        print(df[df.duplicated(['cumsum'], keep=False)])
    else:
        dict_to_transform = df['cumsum'].to_dict()
#        print(dict_to_transform)
        dataframe_to_transform[column] = dataframe_to_transform[column].replace(dict_to_transform)
#        print(dataframe_to_transform)
```
