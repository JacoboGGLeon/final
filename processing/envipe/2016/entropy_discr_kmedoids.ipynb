## NUMERICAL VARIABLES

### manhattan weighted by rank 
`mrw`

```r
data_entropy <- as.matrix(data_entropy)
dunn_var_max_entropy <- c()



for (row in 1:nrow(data_entropy)) {
    print(paste('evaluando: ', row, data_entropy[row]))
    
    if(row > 1){
        #append columns
        data <- cbind(data, data_transform[data_entropy[row]])
        
#        data <- as.matrix(data)
        dist_data <- kmed::distNumeric(x = data,
                                       y = data,
                                       method = "mrw")

        range_k = 2:10
        dunn_kmedoids <- c()
        set.seed(123)

        for (k in range_k) {
          set.seed(123)

#          kmeans <- stats::kmeans(x = data, centers = k, nstart = 25)
            kmedoids <- kmed::fastkmed(distdata = dist_data,
                                       ncluster = k, 
                                       iterate = 25)

          index_internal_kmeans <- clusterCrit::intCriteria(traj = data,
                                                            part = as.integer(kmedoids$cluster), 
                                                            #part = kmeans$cluster, 
                                                            crit = "dunn")
          dunn_kmeans[k-1] <- index_internal_kmeans$dunn
#          dunn_kmeans[k-2] <- index_internal_kmeans$dunn
            print(paste('k:', k, 'dunn value:', index_internal_kmeans$dunn))
        }
        
        dunn_var_max_entropy[row] <- max(dunn_kmeans)
        print(paste('max dunn value:', max(dunn_kmeans), 'in', 'k:',which.max(dunn_kmeans)+1))
#        print(paste('max dunn value:', max(dunn_kmeans), 'in', 'k:',which.max(dunn_kmeans)+2))
            
    } else if (row == 1){
        #create a matrix with the first column
        data <- matrix()
        data <- cbind(data, data_transform[data_entropy[row]])
        data$data <- NULL
        
        #0 para el primer valor
        dunn_var_max_entropy[row] <- 0

    }
    
#    print(head(data))
}
```

```r
plot_dunn_entropy <- ggplot2::qplot(x = c(1:length(dunn_var_max_entropy)), 
                                   y = dunn_var_max_entropy,
                                   geom = 'line') 
#                                   main = 'k-means', 
#                                   xlab = 'k', 
#                                   ylab = 'Distance Ratio')

plot_dunn_entropy <- plot_dunn_entropy + 
theme_bw() +
geom_vline(xintercept = c(1:length(dunn_var_max_entropy))[which.max(dunn_var_max_entropy)],
             linetype = 2, color='red')   
#ggsave(filename = "../../images/enh/dunn_kmeans_2_15.png", dpi = 300)

plot_dunn_entropy
```

```r
vect_best__entropy <- data_entropy[1:c(1:length(dunn_var_max_entropy))[which.max(dunn_var_max_entropy)]]
vect_best__entropy
```

```r
head(data_transform[vect_best__entropy])
```

```r
data <- as.matrix(data_transform[vect_best__entropy])

range_k = 2:10
dunn_kmeans <- c()
set.seed(123)

for (k in range_k) {
  set.seed(123)
  
  kmeans <- stats::kmeans(x = data, centers = k, nstart = 25)
  
  index_internal_kmeans <- clusterCrit::intCriteria(traj = data,
                                                    part = as.integer(kmeans$cluster), 
                                                    #part = kmeans$cluster, 
                                                    crit = "dunn")
  dunn_kmeans[k-1] <- index_internal_kmeans$dunn
}
```

```r
plot_dunn_kmeans <- ggplot2::qplot(x = range_k, 
                                   y = dunn_kmeans, 
                                   geom = 'line', 
                                   main = 'k-means', 
                                   xlab = 'k', 
                                   ylab = 'Distance Ratio')

plot_dunn_kmeans <- plot_dunn_kmeans  + 
  theme_bw() + #ylim(0,0.3) + 
  geom_vline(xintercept = range_k[which.max(dunn_kmeans)],
             linetype = 2, color='red')   
#ggsave(filename = "../../images/enh/dunn_kmeans_2_15.png", dpi = 300)

plot_dunn_kmeans
```

```r
kmeans <- stats::kmeans(x = data, 
                        centers = range_k[which.max(dunn_kmeans)], 
                        nstart = 25)

plot_kmeans <- factoextra::fviz_cluster(object = kmeans,
                                        data = data,
                                        stand = F, 
                                        geom =  'point', 
                                        ellipse.type = 'convex',
                                        ellipse = T,
                                        main = paste('Clustering: k-means k:', range_k[which.max(dunn_kmeans)]),
                                        outlier.color = 'black',
                                        show.clust.cent = T)

plot_kmeans <- plot_kmeans + 
  theme_bw() +
  theme(legend.position = 'bottom')  

plot_kmeans
```
