## Módulo sobre victimización 

`tmod_vic_envipe2016`

```python
file_tmod_vic_envipe2016 = '../../../data/inegi-encuestas/envipe/envipe2018_csv/tmod_vic_envipe_2018/conjunto_de_datos/conjunto_de_datos_tmod_vic_envipe_2018.csv'
```

```python
df_tmod_vic_envipe2016 = pd.read_csv(file_tmod_vic_envipe2016, encoding='latin')
```

```python
df_tmod_vic_envipe2016_copy = df_tmod_vic_envipe2016.copy()
```

```python
df_tmod_vic_envipe2016_copy.head()
```

```python
df_merge_tmod_vic_envipe2016_copy = pd.merge(left=df_tmod_vic_envipe2016_copy, 
                                             right=df_tvivienda_envipe2016_copy[['ID_VIV', 'CVE_ENT', 'NOM_ENT', 'CVE_MUN', 'NOM_MUN']],
                                             how='inner', 
                                             on=['ID_VIV'])
```

```python
df_merge_tmod_vic_envipe2016_copy.columns
```

### Subset Ciudad de México

```python
df_tmod_vic_envipe2016_09 = df_merge_tmod_vic_envipe2016_copy[df_merge_tmod_vic_envipe2016_copy.CVE_ENT == 9]
```

```python
df_tmod_vic_envipe2016_09.head()
```

```python
msno.matrix(df_tmod_vic_envipe2016_09, fontsize=50)
plt.xlabel('Columnas', fontsize=30)
plt.ylabel('Observaciones', fontsize=30)
plt.savefig('../../../images/envipe/df_tmod_vic_envipe2018_09.png', dpi=300)
```

```python
df_tmod_vic_envipe2016_09.replace(to_replace='\r',
                                  value=np.NaN,
                                  inplace=True)
```

```python
msno.matrix(df_tmod_vic_envipe2016_09, fontsize=50)
plt.xlabel('Columnas', fontsize=30)
plt.ylabel('Observaciones', fontsize=30)
plt.savefig('../../../images/envipe/df_tmod_vic_envipe2018_09_nan.png', dpi=300)
```

```python
#list(df_tmod_vic_envipe2016_09.columns)
```

```python
list_identificadores = ['ID_VIV', 'ID_HOG', 'ID_PER', 'ID_DEL', 'UPM', 'VIV_SEL', 'HOGAR', 'ND_TIPO', 'TD_TIPO', 'R_SEL', 'RESUL_H', 'AREAM_OCU','FAC_DEL', 'FAC_DEL_AM', 'DOMINIO', 'ESTRATO', 'EST_DIS', 'UPM_DIS']
```

```python
df_tmod_vic_envipe2016_09.drop(labels=list_identificadores, inplace=True, axis=1)
```

```python
msno.bar(df_tmod_vic_envipe2016_09, fontsize=10)
plt.xlabel('Columnas', fontsize=30)
plt.ylabel('Observaciones', fontsize=30)
plt.savefig('../../../images/envipe/df_tmod_vic_envipe2018_09_nan_ident.png', dpi=300)
```

```python
df_tmod_vic_envipe2016_09.fillna(value=-1, inplace=True)
```

```python
threshold = round(df_tmod_vic_envipe2016_09.shape[0]*.8)
threshold
```

```python
l = []
for column in df_tmod_vic_envipe2016_09.columns:
    na_column =  df_tmod_vic_envipe2016_09[column].value_counts().get(-1)
    if(na_column != None):
        print(column, ': ', df_tmod_vic_envipe2016_09.shape[0] - na_column, ' - ', threshold)
        if (df_tmod_vic_envipe2016_09.shape[0] - na_column < threshold):
            print(column, 'd e l e t e d')
            l.append(column)
```

```python
df_tmod_vic_envipe2016_09.drop(labels=l, inplace=True, axis=1)
```

```python
df_tmod_vic_envipe2016_09.replace(to_replace=-1,
                                  value=np.NaN,
                                  inplace=True)
```

```python
msno.bar(df_tmod_vic_envipe2016_09, fontsize=10)
plt.xlabel('Columnas', fontsize=30)
plt.ylabel('Observaciones', fontsize=30)
plt.savefig('../../../images/envipe/df_tmod_vic_envipe2018_09_nan_threshold.png', dpi=300)
```

```python
l = []
for column in df_tmod_vic_envipe2016_09.columns:
    try:
        df_tmod_vic_envipe2016_09[column] = df_tmod_vic_envipe2016_09[column].astype('int64')
    except:
        l.append(column)
    print('COLUMNA: ', column)
    print('NUNIQUE: ', df_tmod_vic_envipe2016_09[column].nunique())
    print('UNIQUE: ', df_tmod_vic_envipe2016_09[column].unique())
    print('VALUE COUNTS ', df_tmod_vic_envipe2016_09[column].value_counts())
    print('___________________________________')
    
```

```python
file_out_tmod_vic_envipe2016 = '../../../dataframes/envipe/tmod_vic_envipe2018_09.csv'
```

```python
df_tmod_vic_envipe2016_09.to_csv(file_out_tmod_vic_envipe2016)
```
