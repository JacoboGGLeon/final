```python
file = '../../../data/inegi-localidades/AGEEML_20193181542178.csv'
```

```python
df = pd.read_csv(file, encoding = "latin")
```

```python
df_loc_copy = df.copy()
```

```python
df_loc_copy.head()
```

```python
delegaciones = df_loc_copy[['CVE_MUN','NOM_MUN']].copy()
delegaciones
```

```python
df_copy.rename(columns={'Alcaldía':'NOM_MUN'}, inplace=True)
```

```python
df_sec_deleg = pd.merge(left=df_copy, right=delegaciones, on='NOM_MUN')
df_sec_deleg
```

```python
df_sec_deleg.CVE_MUN.value_counts()
```

```python
file = '../../../dataframes/inegi/df_sec_deleg.csv'
```

```python
df_sec_deleg.to_csv(file, index=False)
```
